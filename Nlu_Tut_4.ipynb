# **Question Classifier 50 class**
50 Class Questions Classifier example

Classifies between 50 different types of questions trained on the Trec50 dataset When setting predict(meta=True) nlu will output the probabilities for all other 49 question classes. 

```python
pipe=nlu.load('en.classify.trec50')
pipe.predict('Due to the skillset developed, I was offered a job even before the program ended')
```
