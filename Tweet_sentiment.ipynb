```python
#remove emojios and emoticons
import re

clean_txt=[]
def deEmojify(text):
    regrex_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002500-\U00002BEF"  # chinese char
                               u"\U00002702-\U000027B0"
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               u"\U0001f926-\U0001f937"
                               u"\U00010000-\U0010ffff"
                               u"\u2640-\u2642"
                               u"\u2600-\u2B55"
                               u"\u200d"
                               u"\u23cf"
                               u"\u23e9"
                               u"\u231a"
                               u"\ufe0f"  # dingbats
                               u"\u3030"
                               "]+", flags=re.UNICODE)
    return regrex_pattern.sub(r'',text)

for tx in txt:
    clean_txt.append(deEmojify(tx)) 
#print(emoji_pattern.sub(r'', text)) # no emoji
print(clean_txt)
```

```python
cl_text=[]
for text in clean_txt:
    text=re.findall("[a-zA-Z]+", text)
    #text = re.sub(r'^https?:\/\/.*[\r\n]*', '', text, flags=re.MULTILINE)
    cl_text.append(text)
print(cl_text)
```

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
import numpy as np
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
stop_words = set(stopwords.words('english'))
for txt in retained:
    for text in txt:
        tokenized = sent_tokenize(text)
        for i in tokenized:
            # Word tokenizers is used to find the words
            # and punctuation in a string
            wordsList = nltk.word_tokenize(i)
            # removing stop words from wordList
            wordsList = [w for w in wordsList if not w in stop_words]
            # Using a Tagger. Which is part-of-speech
            # tagger or POS-tagger.
            tagged = list(nltk.pos_tag(wordsList))
            print(np.shape(tagged))
            tg=np.array(tagged)
            print(tg)
```

```python
f1=open('negative-words.txt')
f2=open('positive-words.txt')
nw=f1.read()
pw=f2.read()
#print(pw)
ncount=0
pcount=0
polarit=0
wc=0
pl=[]
print(np.shape(Tdf))
for i in range(len(Tdf)):
    ncount=0
    pcount=0
    polarit=0
    count=0
    for index, df in Tdf.iterrows():
        df=re.findall("[a-zA-Z]+", str(df))
        for tx in str(df).split():
            count=count+1
            txt=""
            for ch in tx:
                if ch=="'" or ch==']' or ch=='[' or ch==',' or ch=='Tweet' or ch=='Text':
                    continue
                else:
                    txt=txt+ch
            txt=str(txt)
            #print(txt, np.size(txt),len(txt),type(txt))
            if txt.lower() in nw.split():
                #print(txt)
                #print("negative")
                ncount=ncount+1
            elif txt.lower() in pw.split():
                #print("positive")
                pcount=pcount+1
            #else:
                #print("Neutral")
           #print(ncount,pcount,wc)   
        polarity= ((pcount-ncount)/(count))
        count=0
        if polarity<2:
            pl.append(-1)
        elif polarity>=2 and polarity<=4:
            pl.append(0)
        else:
            pl.append(1)
    Tdf["Polarity"]=pd.DataFrame(pl)
```
