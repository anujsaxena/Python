```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
```

```python
Data = [[[i+j] for i in range(5)] for j in range(100)]
print(np.array(Data).shape)
```

```python
print(Data)
```

```python
target =[[i+3] for i in range(100)]
```

```python
print(np.array(target).shape)
```

```python
data = np.array(Data)
target = np.array(target)
```

```python
data
```

```python
data.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.2, random_state=9)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
model=Sequential()
```

```python
model.add(LSTM(1, batch_input_shape=(None, 5,1), return_state=False))
```

```python
model.compile(loss='mean_absolute_error', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X_train, y_train, epochs=5, validation_data=(X_test, y_test))
```

```python
results= model.predict(X_test)
```

```python
results
```

```python
plt.scatter(range(20), results, c='r')
plt.scatter(range(20), y_test, c='g')
plt.show()
```

```python
model_output=model.fit(X_train, y_train, epochs=50, validation_data=(X_test, y_test))
```

```python
plt.plot(model_output.history['loss'])
plt.show()
```

```python
Data =[[[(i+j)/100] for i in range(5)] for j in range(100)]
target = [[(i+2)/100] for i in range(100)]
```

```python
data = np.array(Data)
target= np.array(target)
X_train, X_test, Y_train, Y_test = train_test_split(data, target, test_size=0.2, random_state=6)
model= Sequential()
```

```python
model.add(LSTM(1, batch_input_shape=(None, None, 1), return_sequences=True))
model.add(LSTM(1, return_sequences=False))
```

```python
model.compile(loss='mean_absolute_error', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
model_output=model.fit(X_train, Y_train, epochs=1200, validation_data=(X_test, Y_test))
```

```python
results = model.predict(X_test)
plt.scatter(range(20), results, c='r')
plt.scatter(range(20), Y_test, c='g')
plt.show()
```
